```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb
```

```python
%matplotlib inline

rcParams['figure.figsize'] = 5, 4
sb.set_style('whitegrid')
```

### Ploting a line chart in matplotlib

```python
x  = range(1, 10)
y = [1,2,3,4,0,4,3,2,1]

plt.plot(x,y)
```

### plotting a line chart from a Pandas object

```python
filename = 'data/mtcars.csv'
cars = pd.read_csv(filename)
columns = list(cars.columns.values)
print(columns)
columns[0] = 'car names'
print(columns)
cars.columns = columns
cars.head()
```

```python
mpg = cars['mpg']
mpg.plot()
```

```python
df = cars[['cyl', 'wt', 'mpg']]
df.plot()
```

## Creating bar Charts

### Creating a bar chart from a list

```python
plt.bar(x, y)
```

### Creating bar charts from Pandas objects

```python
mpg.plot(kind='bar')
```

```python
mpg.plot(kind='barh')
```

## Creating a pie chart

```python
x = [1,2,3,4,5]
plt.pie(x)
```

## Saving a plot

```python
plt.savefig('pie_chart.jpg', dpi=100)
plt.show()
```
