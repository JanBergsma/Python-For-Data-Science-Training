```python
def plotyear(year):
    data = gapminder[gapminder.year == year]
    area = 5e-6 * data.population
    colors = data.region.map({
        'Africa': 'skyblue',
        'Europe': 'gold',
        'America': 'palegreen',
        'Asia': 'Coral'
        ''
    })
    data.plot.scatter(
        'babies_per_woman', 
        'age5_surviving',
        s=area,
        color=colors,
        linewidth=1,
        edgecolors='k',
        figsize=(12, 9)
    )
    pp.axis(ymin=50, ymax=105, xmin=0, xmax=8)
    pp.xlabel('babies per woman')
    pp.ylabel('% children alive at 5')
```

```python
plotyear(1965)
```

```python
interact(plotyear, year=widgets.IntSlider(min=1950, max=2015, step=1, values=1965))
```
