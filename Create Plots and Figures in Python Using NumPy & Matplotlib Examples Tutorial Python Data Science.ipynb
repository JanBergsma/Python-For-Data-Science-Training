# Create Plots and Figures in Python Using NumPy & Matplotlib Examples Tutorial Python Data Science

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = [10, 10]
```

```python
# create normaly distibuted data
mu, sigma = 100, 15
data_set = mu + sigma * np.random.randn(10_000)
```

```python
# the histogram of the data
n, bins, patches = plt.hist(
    data_set, 
    50, 
    density=True, 
    facecolor='#ff1493',
    alpha=0.75
)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, 0.025, r'$\mu=100,\ \sigma-15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

## Multiple Lines, Single Plot

### Random low and high temperature data

```python
data_set_size = 15
low_mu, low_sigma = 50, 4.3
low_data_set = low_mu + low_sigma * np.random.randn(data_set_size)

high_mu, high_sigma = 57, 5.2
high_data_set = high_mu + high_sigma * np.random.randn(data_set_size)

days = list(range(1, data_set_size + 1))
```

```python
plt.plot(days, low_data_set)
```

```python
plt.plot(days, low_data_set, high_data_set)
```

```python
plt.plot(
    days, low_data_set,
    days, low_data_set, 'vb',
    days, high_data_set,
    days, high_data_set, '^y'
)
plt.show()
```

```python
plt.plot(
    days, low_data_set,
    days, low_data_set, 'vb',
    days, high_data_set,
    days, high_data_set, '^y'
)
plt.xlabel('Day')
plt.ylabel('Temperature: degrees Fahrenheit')
plt.title('Randomized temperature data')
plt.show()
```

```python
plt.plot(
    days, high_data_set, '^k'
)
plt.xlabel('Day')
plt.ylabel('Temperature: degrees Fahrenheit')
plt.title('Randomized temperature data')
plt.show()
```

```python
t1 = np.arange(0.0, 2.0, 0.1)
t2 = np.arange(0.0, 2.0, 0.01)

# note that plot returns a list of lines.  The "l1, = plot" usage
# extracts the first element of the list into l1 using tuple
# unpacking.  So l1 is a Line2D instance, not a sequence of lines
l1, = plt.plot(t2, np.exp(-t2))
l2, l3 = plt.plot(t2, np.sin(2 * np.pi * t2), '--o', t1, np.log(1 + t1), '.')
l4, = plt.plot(t2, np.exp(-t2) * np.sin(2 * np.pi * t2), 's-.')

plt.legend((l2, l4), ('oscillatory', 'damped'), loc='upper right', shadow=True)
plt.xlabel('time')
plt.ylabel('volts')
plt.title('Damped oscillation')
plt.show()
```
