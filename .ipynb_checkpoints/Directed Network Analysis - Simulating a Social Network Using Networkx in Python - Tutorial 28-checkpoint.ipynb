## Simulating a Social network (Directed Metwork Analysis

```python
import numpy as np
import pandas as nx

import networkx as nx

from pylab import rcParams
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

rcParams['figure.figsize'] = 10, 8
sb.set_style('whitegrid')
```

## Generating a graph object and edgelist

```python
DG = nx.gn_graph(7, seed=25)
for line in nx.generate_edgelist(DG, data=False):
    print(line)
```

```python
print(DG.nodes[0])
```

## Assigning attributes to nodes

```python
DG.nodes[0]['name'] = 'Alice'
DG.nodes[1]['name'] = 'Bob'
DG.nodes[2]['name'] = 'Claire'
DG.nodes[3]['name'] = 'Dennis'
DG.nodes[4]['name'] = 'Esther'
DG.nodes[5]['name'] = 'Frank'
DG.nodes[6]['name'] = 'George'
```

```python
[DG.nodes[i] for i in range(len(DG.nodes))]
```

```python
DG.add_nodes_from([
    (0, {'age': 25, 'gender': 'f'}),
    (1, {'age': 31, 'gender': 'm'}),
    (2, {'age': 18, 'gender': 'f'}),
    (3, {'age': 47, 'gender': 'm'}),
    (4, {'age': 22, 'gender': 'f'}),
    (5, {'age': 23, 'gender': 'm'}),
    (6, {'age': 50, 'gender': 'g'}),
])
```

```python
[DG.nodes[i] for i in range(len(DG.nodes))]
```

## Visualize your network graph

```python
nx.draw_circular(DG, node_color='bisque', with_labels=True)
```

```python
label_nodes = {i: DG.nodes[i]['name'] for i in DG.nodes}
nx.draw_circular(
    DG, 
    labels=label_nodes,
    node_color='bisque', 
    with_labels=True,
)
```

```python
G = DG.to_directed()
```

```python
nx.draw_spectral(
    G, 
    labels=label_nodes,
    node_color='bisque', 
    with_labels=True,
)
```
