# Histogram, Box plots and Scatter plots using Seaborn and Matplotlib in Python - Tutorial 11

## Constructing histograms, box plots, and scatter plots

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pandas.plotting import scatter_matrix

from matplotlib import rcParams
import seaborn as sb
```

```python
%matplotlib inline

rcParams['figure.figsize'] = 10, 8
sb.set_style('whitegrid')
```

### Eyeballing dataset distributions with histograms

```python
uri = 'data/mtcars.csv'
cars = pd.read_csv(uri)
columns = list(cars.columns.values)
columns[0] = 'car_names'
cars.columns = columns
cars.index = cars.car_names
cars.head()
```

```python
mpg = cars['mpg']
mpg.plot(kind='hist')
```

```python
mpg.hist()
```

plt.hist(mpg)

```python
sb.distplot(mpg)
```

### Seeing scatterplots in action

```python
cars.plot(kind='scatter', x='hp', y='mpg', c=['darkgray'], s=150)
```

```python
sb.regplot(x='hp', y='mpg', data=cars, scatter=True)
```

### Generating a scatter plot matrix

```python
sb.pairplot(cars)
```

```python
cars_df = pd.DataFrame(
    cars.iloc[:, cars.columns.get_indexer(['mpg', 'disp', 'hp', 'wt'])].values,
    columns=['mpg', 'disp', 'hp', 'wt']
)
cars_target = cars.iloc[:, 9].values
cars_df['group'] = pd.Series(cars_target, dtype='category')

sb.pairplot(cars_df, hue='group', palette='hls', diag_kind='hist')
```

### Building boxplots

```python
cars.boxplot(column='mpg', by='am')
cars.boxplot(column='wt', by='am')
```

```python
sb.boxplot(x='am', y='mpg', data=cars, palette='hls')
```
