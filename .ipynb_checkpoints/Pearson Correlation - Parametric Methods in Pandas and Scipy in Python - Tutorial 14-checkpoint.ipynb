# Starting with parametric methodes in pandas and scipy

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt

from pylab import rcParams
import seaborn as sb

import scipy as sc
from scipy.stats import pearsonr
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 15, 12
sb.set_style('whitegrid')
```

### The Pearson Correlation

```python
uri = 'data/mtcars.csv'
cars = pd.read_csv(uri)
columns = list(cars.columns.values)
columns[0] = 'car_names'
cars.columns = columns
cars.index = cars.car_names
cars.head()
```

```python
sb.pairplot(cars)
```

```python
X = cars[['mpg', 'hp', 'qsec', 'wt']]
sb.pairplot(X)
```

### Using scipy to calculate the Pearson correlation coeeficient

```python
mpg = cars['mpg']
hp = cars['hp']
qsec = cars['qsec']
wt = cars['wt']
```

```python
pearsonr_coefficient, p_value = pearsonr(mpg, hp)
print('mpg - hp')
print(f'Pearson R Correlation Coefficient {pearsonr_coefficient:0.3}')
print(f'Two-tailed p-value {p_value:0.3}')
```

```python
pearsonr_coefficient, p_value = pearsonr(mpg, qsec)
print('mpg - qseq')
print(f'Pearson R Correlation Coefficient {pearsonr_coefficient:0.3}')
print(f'Two-tailed p-value {p_value:0.3}')
```

```python
pearsonr_coefficient, p_value = pearsonr(mpg, wt)
print('mpg - wt')
print(f'Pearson R Correlation Coefficient {pearsonr_coefficient:0.3}')
print(f'Two-tailed p-value {p_value:0.3}')
```

```python
pearsonr_coefficient, p_value = pearsonr(hp, qsec)
print('hp - qsec')
print(f'Pearson R Correlation Coefficient {pearsonr_coefficient:0.3}')
print(f'Two-tailed p-value {p_value:0.3}')
```

```python
pearsonr_coefficient, p_value = pearsonr(hp, wt)
print('hp - wt')
print(f'Pearson R Correlation Coefficient {pearsonr_coefficient:0.3}')
print(f'Two-tailed p-value {p_value:0.3}')
```

```python
pearsonr_coefficient, p_value = pearsonr(qsec, wt)
print('qsec - wt')
print(f'Pearson R Correlation Coefficient {pearsonr_coefficient:0.3}')
print(f'Two-tailed p-value {p_value:0.3}')
```

### Using pandas to calculate the Pearson correlation coeeficient

```python
corr = X.corr()
corr
```

### Using Seaborn to visualize the Pearson correlation coefficient

```python
sb.heatmap(
    corr, 
    xticklabels=corr.columns.values, 
    yticklabels=corr.columns.values,
    cmap="coolwarm"
)
```
