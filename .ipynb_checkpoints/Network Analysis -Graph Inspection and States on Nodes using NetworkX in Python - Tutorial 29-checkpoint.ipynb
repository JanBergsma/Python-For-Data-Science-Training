## Analyzing a Social Metwork

```python
import numpy as np
import pandas as nx

import networkx as nx

from pylab import rcParams
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
%matplotlib inline

rcParams['figure.figsize'] = 10, 8
sb.set_style('whitegrid')
```

```python
DG = nx.gn_graph(7, seed=25)

DG.add_nodes_from([
    (0, {'name': 'Alice', 'age': 25, 'gender': 'f'}),
    (1, {'name': 'Bob', 'age': 31, 'gender': 'm'}),
    (2, {'name': 'Claire', 'age': 18, 'gender': 'f'}),
    (3, {'name': 'Dennis', 'age': 47, 'gender': 'm'}),
    (4, {'name': 'Esther', 'age': 22, 'gender': 'f'}),
    (5, {'name': 'Frank', 'age': 23, 'gender': 'm'}),
    (6, {'name': 'George', 'age': 50, 'gender': 'g'}),
])
```

```python
G = DG.to_undirected()
```

```python
print(nx.info(DG))
print('-'*40)
print(nx.info(G))
```

## Considering degrees in a social network

```python
DG.degree()
```

## Identifying successor nodes

```python
# label_nodes = {i: DG.nodes[i]['name'] for i in DG.nodes}
nx.draw_circular(
    DG, 
#     labels=label_nodes,
    node_color='bisque', 
    with_labels=True,
)
```

```python
list(DG.successors(3))
```

```python
list(DG.successors(4))
```
