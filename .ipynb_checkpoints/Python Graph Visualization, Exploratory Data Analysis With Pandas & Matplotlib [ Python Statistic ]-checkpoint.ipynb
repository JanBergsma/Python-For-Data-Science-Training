```python
gapminder = pd.read_csv(filename)
```

```python
def plotyear(year):
    data = gapminder[gapminder.year == year]
    data.plot.scatter('gdp_per_day', 'life_expectancy', 
                      logx=True, figsize=(12, 9)
    )
    pp.axis(xmin=1, xmax=200)
```

```python
plotyear(1965)
```

```python
def plotyear(year):
    data = gapminder[gapminder.year == year].sort_values('population', ascending=True)
    area = 5e-6 * data.population
    
    color = data.age5_surviving
    
    data.plot.scatter(
        'gdp_per_day', 
        'life_expectancy', 
        logx=True,
        figsize=(12, 9),
        s=area,
        linewidth=1,
        edgecolor='k',
        c = color,
        colormap = matplotlib.cm.get_cmap('Purples_r'),
        vmin=55, vmax=100,
        sharex=False
    )
    pp.axis(xmin=1, xmax=200)
    
plotyear(1965)
```

```python
def plotyear(year):
    data = gapminder[gapminder.year == year].sort_values('population', ascending=True)
    area = 5e-6 * data.population
    
    color = data.age5_surviving
    edgecolor = data.region.map({
        'Africa': 'skyblue',
        'Europe': 'gold',
        'America': 'palegreen',
        'Asia': 'Coral'
        ''
    })

    data.plot.scatter(
        'gdp_per_day', 
        'life_expectancy', 
        logx=True,
        figsize=(10, 9),
        s=area,
        linewidth=1,
        edgecolor=edgecolor,
        c = color,
        colormap = matplotlib.cm.get_cmap('Purples_r'),
        vmin=55, vmax=100,
        sharex=False
    )
    pp.axis(xmin=1, xmax=500, ymin=30, ymax=100)
    
plotyear(1965)
```

```python
interact(plotyear, year=range(1965, 2016, 10))
```

```python
def plotyear(year):
    data = gapminder[gapminder.year == year].sort_values('population', ascending=True)
    area = 5e-6 * data.population
    
    color = data.age5_surviving
    edgecolor = data.region.map({
        'Africa': 'skyblue',
        'Europe': 'gold',
        'America': 'palegreen',
        'Asia': 'Coral'
        ''
    })

    data.plot.scatter(
        'gdp_per_day', 
        'life_expectancy', 
        logx=True,
        figsize=(10, 9),
        s=area,
        linewidth=1,
        edgecolor=edgecolor,
        c = color,
        colormap = matplotlib.cm.get_cmap('Purples_r'),
        vmin=55, vmax=100,
        sharex=False
    )
    
    for level in [4, 16, 64]:
        pp.axvline(level, linestyle=':', color='k')
    
    pp.axis(xmin=1, xmax=500, ymin=30, ymax=100)
    
plotyear(1965)
```

```python
plotyear(2015)
```
