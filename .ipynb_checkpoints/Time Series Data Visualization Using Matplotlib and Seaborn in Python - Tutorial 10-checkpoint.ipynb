# Time Series Data Visualization Using Matplotlib and Seaborn in Python - Tutorial 10

## Creating visualizations from time series data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb
```

```python
%matplotlib inline

rcParams['figure.figsize'] = 15, 12
sb.set_style('whitegrid')
```

## The simplest time series plot

```python
filename = 'data/Superstore-Sales.csv'
df = pd.read_csv(filename, index_col='Order Date', parse_dates=True)
df.head()
```

```python
df['Order Quantity'].plot()
```

```python
df2 = df.sample(n=100, random_state=25, axis=0)
plt.xlabel('Order Data')
plt.ylabel('Order Quantity')
plt.title('Superstore Sales')

df2['Order Quantity'].plot()
```
