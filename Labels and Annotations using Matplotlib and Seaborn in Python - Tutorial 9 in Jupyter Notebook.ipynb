# Labels and Annotations using Matplotlib and Seaborn in Python - Tutorial 9 in Jupyter Notebook

## Creating labels and annotations

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import seaborn as sb
```

```python
%matplotlib inline

rcParams['figure.figsize'] = 15, 12
sb.set_style('whitegrid')
```

## Labelling plot features

### The functional method

```python
x = range(1, 10)
y = [1,2,3,4,5,4,3,2,1]

plt.bar(x, y)

plt.xlabel( 'your x-axis label')
plt.ylabel( 'your y-axis label')
plt.show()
```

```python
z = [1,2,3,4,5]

veh_type = ['bicycle', 'moterbike', 'car', 'van', 'stroller']
plt.pie(z, labels=veh_type)
plt.show()
```

### The object-oriented method

```python
uri = 'data/mtcars.csv'
cars = pd.read_csv(uri)
columns = list(cars.columns.values)
columns[0] = 'car_names'
cars.columns = columns
cars.head()
```

```python
mpg = cars.mpg

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Mile per Gallon of Cars in Mtcars')
ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')
```

## Adding a legend to your plot

### The functional method

```python
plt.pie(z)

plt.legend(veh_type, loc='best')
plt.show()
```

### The object-oriented method

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Mile per Gallon of Cars in Mtcars')
ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')

ax.legend(loc='best')
```

## Anotating your plot

```python
mpg.max()
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Mile per Gallon of Cars in Mtcars')
ax.set_xlabel('car names')
ax.set_ylabel('miles/gal')

ax.legend(loc='best')
ax.set_ylim([0,45])
ax.annotate(
    'Toyota Corolla', 
    xy=(19,33.9), 
    xytext=(21,35), 
    arrowprops=dict(facecolor = 'black', shrink = 0.05)
)
```
